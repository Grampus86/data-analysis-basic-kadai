```python
import numpy as np
import matplotlib.pyplot as plt

#折れ線グラフ
y0 = [1, 2, -5, 2]

# sinカーブ
x1 = np.linspace(0, 10, 100)
y1 = 2 + 2 * np.sin(2 * x1)

# ヒストグラム
y2 = np.random.normal(15, 5, 2000)

# 散布図
np.random.seed(3)
x3 = 4 + np.random.normal(0, 2, 60)
y3 = 4 + np.random.normal(0, 2, len(x3))

# プロット
fig, axes = plt.subplots(nrows=2, ncols=3, tight_layout=True)

axes[0, 0].plot(y0) # 折れ線
axes[0, 1].plot(y1) # sinカーブ
axes[0, 2].hist(y2) # ヒスト
axes[1, 0].scatter(x3, y3) # 散布図

plt.show()
```
